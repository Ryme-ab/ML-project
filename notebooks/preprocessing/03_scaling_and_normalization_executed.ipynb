# 03 – Scaling & Normalization

**Objective:** Transform numeric features into comparable scales so that distance-based algorithms (clustering, k-NN, neural networks, SVMs) and gradient-based models behave correctly.

**Strategies applied:**
- **StandardScaler** (z-score: mean=0, std=1) – best for normally distributed features.
- **MinMaxScaler** (range [0, 1]) – best when bounded outputs are needed (e.g., neural nets).
- **RobustScaler** (median & IQR) – best when outliers are present (trade data often has extreme values).

**Input:** `data/processed/02_*.csv`
**Output:** `data/processed/03_scaled_*.csv` with original + scaled columns.

## 3.1 – Setup

```python
import numpy as np
import pandas as pd
import warnings
from pathlib import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 50)

PROJECT_ROOT = Path.cwd().parent.parent
PROCESSED_DIR = PROJECT_ROOT / 'data' / 'processed'

# Load feature-engineered datasets
datasets = {}
for fname in ['02_comtrade_features.csv', '02_worldbank_features.csv',
              '02_wto_demand_features.csv', '02_trade_balance.csv',
              '02_diversification_metrics.csv']:
    path = PROCESSED_DIR / fname
    if path.exists():
        datasets[fname.replace('.csv', '')] = pd.read_csv(path)
        print(f'Loaded {fname}: {datasets[fname.replace(".csv", "")].shape}')
    else:
        print(f'Skipped {fname} (not found)')
```

## 3.2 – Helper Functions

```python
def get_numeric_cols(df, exclude=None):
    """Return numeric columns excluding identifiers and years."""
    if exclude is None:
        exclude = ['year', 'refyear', 'country_code', 'reporteriso', 'partneriso',
                   'cmdcode', 'product_sector_code', 'indicator', 'source', 'flow_type']
    nums = df.select_dtypes(include=[np.number]).columns.tolist()
    return [c for c in nums if c not in exclude and not c.endswith(('_std', '_minmax', '_robust'))]

def scale_dataframe(df, numeric_cols, scaler, suffix):
    """
    Fit-transform selected numeric columns with a sklearn scaler.
    Returns a new dataframe with additional columns suffixed accordingly.
    """
    df_out = df.copy()
    available = [c for c in numeric_cols if c in df_out.columns and df_out[c].notna().any()]
    if not available:
        return df_out

    # Fill NaNs with median before scaling (common practice; original NaNs remain untouched in raw columns)
    fill_vals = df_out[available].median()
    scaled_matrix = scaler.fit_transform(df_out[available].fillna(fill_vals))

    for i, col in enumerate(available):
        df_out[f'{col}_{suffix}'] = scaled_matrix[:, i]
    return df_out

# Instantiate scalers
scalers = {
    'std': StandardScaler(),
    'minmax': MinMaxScaler(),
    'robust': RobustScaler()
}
print('Scalers ready:', list(scalers.keys()))
```

## 3.3 – Scale World Bank Macro Features

```python
key = '02_worldbank_features'
if key in datasets:
    df = datasets[key].copy()
    num_cols = get_numeric_cols(df)
    print('World Bank numeric columns to scale:', num_cols)

    for suffix, scaler in scalers.items():
        df = scale_dataframe(df, num_cols, scaler, suffix)

    datasets[key] = df
    print('Scaled World Bank shape:', df.shape)
    # Show a few scaled columns
    scaled_cols = [c for c in df.columns if c.endswith(('_std', '_minmax', '_robust'))]
    display(df[['country', 'year'] + scaled_cols[:6]].head())
else:
    print('World Bank dataset not available.')
```

## 3.4 – Scale WTO Demand Features

```python
key = '02_wto_demand_features'
if key in datasets and not datasets[key].empty:
    df = datasets[key].copy()
    num_cols = get_numeric_cols(df)
    print('WTO numeric columns to scale:', num_cols)

    for suffix, scaler in scalers.items():
        df = scale_dataframe(df, num_cols, scaler, suffix)

    datasets[key] = df
    print('Scaled WTO demand shape:', df.shape)
    scaled_cols = [c for c in df.columns if c.endswith(('_std', '_minmax', '_robust'))]
    display(df[['year', 'product_sector'] + scaled_cols[:6]].head())
else:
    print('WTO demand dataset not available or empty.')
```

## 3.5 – Scale Trade Balance & Diversification

```python
for key in ['02_trade_balance', '02_diversification_metrics']:
    if key in datasets and not datasets[key].empty:
        df = datasets[key].copy()
        num_cols = get_numeric_cols(df)
        print(f'\n{key} numeric columns:', num_cols)
        for suffix, scaler in scalers.items():
            df = scale_dataframe(df, num_cols, scaler, suffix)
        datasets[key] = df
        print('Scaled shape:', df.shape)
    else:
        print(f'\n{key} not available or empty.')
```

## 3.6 – Scale Comtrade Features

```python
key = '02_comtrade_features'
if key in datasets and not datasets[key].empty:
    df = datasets[key].copy()
    num_cols = get_numeric_cols(df)
    print('Comtrade numeric columns:', num_cols)
    for suffix, scaler in scalers.items():
        df = scale_dataframe(df, num_cols, scaler, suffix)
    datasets[key] = df
    print('Scaled Comtrade shape:', df.shape)
else:
    print('Comtrade features not available.')
```

## 3.7 – Save All Scaled Datasets

```python
output_mapping = {
    '02_worldbank_features': '03_worldbank_scaled.csv',
    '02_wto_demand_features': '03_wto_demand_scaled.csv',
    '02_trade_balance': '03_trade_balance_scaled.csv',
    '02_diversification_metrics': '03_diversification_scaled.csv',
    '02_comtrade_features': '03_comtrade_scaled.csv'
}

for key, out_name in output_mapping.items():
    if key in datasets and not datasets[key].empty:
        out_path = PROCESSED_DIR / out_name
        datasets[key].to_csv(out_path, index=False)
        print(f'Saved {out_name}  ->  shape {datasets[key].shape}')
    else:
        print(f'Skipped {out_name} (source missing)')

print('\nScaling complete. Proceed to 04_master_integration.ipynb.')
```
